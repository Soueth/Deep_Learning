```python
rn = Sequential([
    InputLayer(shape = (30,)),
    Dense(units = 16, activation='relu', kernel_initializer='random_uniform'),
    Dense(units = 16, activation='relu', kernel_initializer='random_uniform'),
    Dense(units=1, activation='sigmoid')
])
```

```python
rn.fit(x_train, y_train, batch_size=10, epochs=100)
```

### **Cross Validation**

```python
import scikeras

from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras import backend as k
```

```python
scikeras.__version__
```

```python
def criar_rede():
    k.clear_session()
    rn = Sequential([
        InputLayer(shape = (30,)),
        Dense(units = 16, activation='relu', kernel_initializer='random_uniform'),
        Dense(units = 16, activation='relu', kernel_initializer='random_uniform'),
        Dense(units=1, activation='sigmoid')
    ])
    otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5)
    rn.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy'])
    return rn
```

```python
rn = KerasClassifier(model= criar_rede, epochs = 100, batch_size=10)
```

```python
results = cross_val_score(estimator=rn, X = x, y = y, cv = 10, scoring='accuracy')
```

```python
# Sem Dropout
results.mean()
```

```python
results.std()
```

### **Dropout**

```python
from tensorflow.keras.layers import Dropout
```

```python
def criar_rede_dropout():
    k.clear_session()
    rn = Sequential([
        InputLayer(shape = (30,)),
        Dense(units = 16, activation='relu', kernel_initializer='random_uniform'),
        Dropout(rate=0.2),
        Dense(units = 16, activation='relu', kernel_initializer='random_uniform'),
        Dropout(rate=0.2),
        Dense(units=1, activation='sigmoid')
    ])
    otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5)
    rn.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy'])
    return rn
```

```python
rn = KerasClassifier(model= criar_rede_dropout, epochs = 100, batch_size=10)
```

```python
results_dropout = cross_val_score(estimator=rn, X = x, y = y, cv = 10, scoring='accuracy')
```

```python
results_dropout
```

```python
results_dropout.mean()
```

```python
results_dropout.std()
```

### **Tuning dos parâmetros**

```python
from sklearn.model_selection import GridSearchCV
```

```python
def criar_rede_cross_val(optimizer, loss, kernel_initializer, activation, neurons):
    k.clear_session()
    rn = Sequential([
        InputLayer(shape = (30,)),
        Dense(units = neurons, activation=activation, kernel_initializer=kernel_initializer),
        Dropout(rate=0.2),
        Dense(units = neurons, activation=activation, kernel_initializer=kernel_initializer),
        Dropout(rate=0.2),
        Dense(units=1, activation='sigmoid')
    ])
    rn.compile(optimizer=optimizer, loss=loss, metrics=['binary_accuracy'])
    return rn
```

```python
rn = KerasClassifier(model=criar_rede_cross_val)
```
