```python
import pandas as pd
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras import utils as np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
base = pd.read_csv('bases/iris.csv')
base
```

```python
x = base.iloc[:, 0:4].values
pd.DataFrame(x)
```

```python
y = base.iloc[:, 4].values
pd.DataFrame(y)
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
y = np_utils.to_categorical(y)
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
model = Sequential()
model.add(InputLayer(shape=(4,)))
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
model.fit(x_train, y_train, batch_size=10, epochs=1000)
```

```python
model.evaluate(x_test, y_test)
```

```python
prevs = model.predict(x_test)
prevs
```

```python
prevs = prevs > 0.5
prevs
```

```python
y_test2 = [np.argmax(t) for t in y_test]
y_test2
```

```python
prevs2 = [np.argmax(t) for t in prevs]
prevs2
```

```python
accuracy_score(y_test2, prevs2)
```

```python
confusion_matrix(y_test2, prevs2)
```

### **Validação cruzada**

```python
from tensorflow.keras import backend as k
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def criar_rede():
    k.clear_session()
    rn = Sequential([
        InputLayer(shape=(4,)),
        Dense(units=4, activation='relu'),
        Dense(units=4, activation='relu'),
        Dense(units=3, activation='softmax')
    ])
    rn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
    return rn
```

```python
rn = KerasClassifier(model=criar_rede, epochs=250, batch_size=10)
results = cross_val_score(estimator=rn, X=x, y=y, cv=10, scoring='accuracy')
```

```python
results.mean()
```

```python
results.std()
```

### **Grid Search**

```python
from tensorflow.keras.layers import Dropout
from sklearn.model_selection import GridSearchCV
```

```python
def criar_rede_grid_search(optimizer, kernel_initializer, activation, neurons, dropout):
    k.clear_session()
    rn = Sequential([
        InputLayer(shape = (4,)),
        Dense(units = neurons, activation=activation, kernel_initializer=kernel_initializer),
        Dropout(rate=dropout),
        Dense(units = neurons, activation=activation, kernel_initializer=kernel_initializer),
        Dropout(rate=dropout),
        Dense(units=3, activation='softmax')
    ])
    rn.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return rn
```

```python
rn = KerasClassifier(model=criar_rede_grid_search, verbose=0)
```

```python
params = {
    'batch_size': [10, 30],
    'epochs': [150, 200],
    'model__optimizer': ['adam', 'RMSprop'] ,
    'model__kernel_initializer': ['random_uniform', 'glorot_normal'],
    'model__activation': ['relu', 'tanh',],
    'model__neurons': [4, 8, 16],
    'model__dropout': [0.2, 0.3],
}
```

```python
gs = GridSearchCV(estimator=rn, param_grid=params, scoring='accuracy', cv=5, verbose=3)
gs.fit(x, y)
```

```python
gs.best_score_, gs.best_params_
```

```python
model = Sequential()
model.add(InputLayer(shape=(4,)))
model.add(Dense(units=16, activation='tanh', kernel_initializer='random_uniform'))
model.add(Dropout(rate=0.2))
model.add(Dense(units=8, activation='tanh', kernel_initializer='random_uniform'))
model.add(Dropout(rate=0.2))
model.add(Dense(units=3, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x, y, batch_size=10, epochs=150)
```

```python
with open('models/iris_struture.json', 'w') as json_file:
    json_file.write(model.to_json())

model.save_weights('models/iris_weights.weights.h5')
model.save('models/iris.h5')
```
