```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from minisom import MiniSom
from sklearn.preprocessing import MinMaxScaler
from matplotlib.pylab import pcolor, colorbar, plot
```

```python
base_train = pd.read_csv('bases/entradas_breast.csv')
base_train
```

```python
base_test = pd.read_csv('bases/saidas_breast.csv')
base_test
```

```python
X = base_train.values
y = base_test.values
```

```python
X
```

```python
y
```

```python
normalizer = MinMaxScaler(feature_range=(0, 1))
X = normalizer.fit_transform(X)
X
```

# **Mapa Auto-organizável**

```python
5 * np.sqrt(X.shape[0])
```

```python
10*12
```

```python
7*17
```

```python
som = MiniSom(x=7, y=17, input_len=X.shape[1], sigma=1.1, learning_rate=0.5, random_seed=2)
```

```python
som.random_weights_init(X)
```

```python
som.train_random(data=X, num_iteration=100000)
```

```python
som._weights
```

```python
som._weights.shape
```

```python
som._activation_map.shape
```

```python
q = som.activation_response(X)
q.shape
```

```python
markers = ['o', 's', 'D']
color = ['r', 'g', 'b']
```

```python
y = y.flatten()
```

```python
pcolor(som.distance_map().T)
colorbar()

for i, x in enumerate(X):
    w = som.winner(x)
    plot(
        w[0] + 0.5, 
        w[1] + 0.5, 
        markers[y[i] - 1], 
        markerfacecolor='None', 
        markeredgecolor=color[y[i]-1], 
        markeredgewidth=2, 
        markersize=10
    )
```
