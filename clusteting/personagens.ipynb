```python
import pandas as pd
import numpy as np

from minisom import MiniSom
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from matplotlib.pylab import pcolor, colorbar, plot
```

```python
base = pd.read_csv('bases/personagens.csv')
base
```

```python
base.isna().sum()
```

```python
X = base.iloc[:, 0:6].values
y = base.iloc[:, 6].values
```

```python
X
```

```python
normalizer = MinMaxScaler(feature_range=(0, 1))
X = normalizer.fit_transform(X)
X
```

```python
y
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

# **Mapa auto-organizÃ¡vel**

```python
5 * np.sqrt(X.shape[0])
```

```python
12*7
```

```python
som = MiniSom(x=12, y=7, input_len=X.shape[1], sigma=1.0, learning_rate=0.5, random_seed=1)
som.random_weights_init(X)
som.train_random(data=X, num_iteration=100)
```

```python
pcolor(som.distance_map().T)
colorbar();

markers = ['o', 's']
colors = ['r', 'g']

for i, x in enumerate(X):
    w = som.winner(x)
    plot(w[0] + 0.5, w[1] + 0.5, markers[y[i]],
         markerfacecolor='None', markersize=10,
         markeredgecolor=colors[y[i]], markeredgewidth=2)
```

```python
mapping = som.win_map(X)
mapping
```

```python
suspects = np.concatenate((mapping[(6, 2)]), axis=0)
suspects = normalizer.inverse_transform(suspects.reshape(-1, 6))
suspects
```

```python
len(base), len(suspects)
```

```python
classes = []

for i in range(len(base)):
    for j in range(len(suspects)):
        if np.all(base.iloc[i, 0:6].values == suspects[j, 0: 6]):
            classes.append(base.iloc[i, 6])

classes = np.asarray(classes)
classes, classes.shape
```

```python
classes
```

```python
final_suspects = np.column_stack((suspects, classes))
final_suspects
```
